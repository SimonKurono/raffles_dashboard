```python
import yfinance as yf

#for TLT vs Raffles
def fetch_and_rebase(ticker, start_date):
    data = yf.download(ticker, start=start_date, interval='1mo')['Close']
    data = data.dropna()
    rebased_data = (data / data.iloc[0]) * 100
    rebased_data.drop(data.index[0], inplace=True)  
    return rebased_data

tlt_data = fetch_and_rebase('TLT', '2002-01-01')
print(tlt_data.head())

```
